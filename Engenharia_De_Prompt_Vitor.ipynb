```python
# -*- coding: utf-8 -*-
"""Trabalho_Engenharia_De_Prompt_Vitor.ipynb

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/1E_4ZyyuHTL4CpbhAkG-SwsEkZ3Kw0SDj
"""

# Instalação das bibliotecas necessárias (caso ainda não estejam instaladas)
!pip install fpdf openai --quiet

import pandas as pd
from openai import OpenAI
from fpdf import FPDF  # biblioteca para gerar PDF
import unicodedata     # biblioteca para sanitizar o texto
from google.colab import files  # biblioteca para permitir download no Colab

# Função para remover caracteres especiais que causam erro no PDF
def limpar_texto(texto):
    return unicodedata.normalize("NFKD", texto).encode("latin-1", "ignore").decode("latin-1")

# Inicialização do cliente com sua chave e base URL do OpenRouter
client = OpenAI(
    api_key="",  # Cole aqui sua chave pessoal
    base_url="https://openrouter.ai/api/v1"
)

# Caminho da sua planilha com os dados
caminho_arquivo = "sheet_prompt_var.xlsx"

# Lê a planilha
df = pd.read_excel(caminho_arquivo)

# Inicializa a estrutura do PDF
pdf = FPDF()
pdf.set_auto_page_break(auto=True, margin=15)
pdf.set_font("Arial", size=12)

# Gera os prompts a partir das variáveis
prompts = []
for index, row in df.iterrows():
    prompt = f"""Você é um especialista em {row['área']}. Monte um Roadmap de {row['tema']} com o objetivo de {row['objetivo']}. Responda em português."""
    prompts.append(prompt)

    # Enviar o prompt para o modelo
    response = client.chat.completions.create(
        model="mistralai/mixtral-8x7b-instruct",
        messages=[{"role": "user", "content": prompt}]
    )

    resposta = response.choices[0].message.content

    # Mostra no console a resposta obtida
    print(f"\n--- Conceito {index + 1}: {row['tema']} ---")
    print(resposta)

    # Adiciona uma nova página ao PDF com o título e conteúdo da resposta
    pdf.add_page()
    pdf.set_font("Arial", "B", 14)
    titulo = f"Roadmap {index + 1}: {row['tema']}"
    pdf.cell(0, 10, limpar_texto(titulo), ln=True)
    pdf.set_font("Arial", size=12)
    pdf.multi_cell(0, 10, limpar_texto(resposta))  # texto sanitizado

# Caminho do arquivo de saída .txt
caminho_saida = "prompts_gerados_python.txt"

# Salva os prompts gerados no arquivo .txt
with open(caminho_saida, "w", encoding="utf-8") as f:
    for prompt in prompts:
        f.write(prompt + "\n\n")

print(f"Prompts salvos em: {caminho_saida}")

# Download automático do .txt no Colab
files.download(caminho_saida)

# Caminho do arquivo PDF final
caminho_pdf = "roadmaps_gerados.pdf"

# Salva o PDF com as respostas da IA
pdf.output(caminho_pdf)

print(f"PDF com roadmaps salvo em: {caminho_pdf}")
```
